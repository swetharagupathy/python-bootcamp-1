```python
weekday = {"Math", "Physics", "English"}
weekend = {"Python", "Math", "AI"}

print(weekday | weekend)
print(weekday & weekend)
print(weekday - weekend)

weekday.add("Biology")
weekend.remove("Math")

print("Python" in weekday or "Python" in weekend)
```

```python
def introduce(name, language="Python"):
    print(f"{name} is learning {language}")

introduce("Swetha")
introduce("Swetha", "SQL")

def calculate_discount(price, discount=10):
    return price - (price * discount / 100)
```

```python
num = int(input())
if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")

score = int(input())
if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
else:
    print("F")

year = int(input())
if (year % 4 == 0 and year % 100 != 0) or year % 400 == 0:
    print("Leap Year")
else:
    print("Not Leap Year")
```
